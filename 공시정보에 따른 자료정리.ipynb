# 유가증권시장, 코스닥시장에 한해서 최근 10년간 공시 당일 일봉부터 일주일 후 까지 보고서 카테고리로 분류

```python
import pandas as pd
```

```python
공시 = pd.read_csv('사용디비.csv',encoding = 'euc-kr')
공시.head()
```

```python
#pandas.query를 이용하여 데이터프레임 분류해주기
#pandas.query로는 데이터프레임 내의 텍스트 데이터만 검색가능 
#코스닥시장과 유가증권시장만을 분류 
data = 공시.query('구분.str.contains("코스닥시장") or 구분.str.contains("유가증권")',engine = 'python')
data.head()
```

```python
#컬럼 확인용
for each in data.columns:
    print(each)
```

### 컬럼 내용정리 

```python
#구분은 시장에 따라 ex 코스닥,유가증권시장 
#CategorySeq는 공시 카테고리에 대한 분류
#보고서명에 공시에 대한 제목 
#내용에는 공시 내용
```

```python
#에러가 발생하여 데이터를 복사해준후 컬럼이름을 수정
#컬럼이름을 앞에 붙인 이유는 filter 를 편하게 하기 위해서 
data2 = data.copy()
data2.rename(columns ={'날짜':'일_날짜', 'CategorySeq':'일_카테고리', '종목명':'일_종목명', '종목코드':'일_종목코드'},inplace = True)
```

```python
data2 = data2.filter(regex = '^일')
data2.head()
```

```python
#r결측값이 있는 행들을 제거 
data2 = data2.dropna()
data2.head()
```

```python
#중복되지 않은 카테고리 확인을 위해
data2['일_카테고리'].unique()
```

```python
#날짜별로 정렬해주기 위해서 
#groupby 이용시 적용해줄 함수 설정 
def top(df):
    return df.sort_values(by='일봉날짜')

#카테고리별로 묶어준 데이터 날짜 순으로 정리.
data2 = data2.groupby('일_카테고리').apply(top)
data2.head()
```

```python
data2['일봉날짜']
```

```python
#날짜 데이터들을 object에서 데이터 형식으로 변경 
data2['일_날짜'] = pd.to_datetime(data2['일_날짜'], format='%Y-%m-%d %H:%M:%S')
data2['일_날짜']
```

```python
#날짜 데이터들을 object에서 데이터 형식으로 변경 
data2['일봉날짜'] = pd.to_datetime(data2['일봉날짜'], format='%Y-%m-%d')
data2['일봉날짜']
```

```python
#데이터양을 조금 줄여주기 위해서 2000년부터의 데이터만 
filter_data = data2.loc[(data2['일봉날짜'] > '2000-01-01')] 
filter_data.head()
```

## 당일의 등락률 구해주기(공시 발표 당일의 저가와 고가를 이용)

```python
당일_등락률 = filter_data.copy()
당일_등락률['당일 등락률'] = (filter_data['일봉고가'] - filter_data['일봉저가'])/filter_data['일봉저가']
당일_등락률 = 당일_등락률.filter(like  = '등락률',axis= 1)
당일_등락률.head()
```

# 공시발표 당일의 시가와 종가를 비교

```python
당일_등락률 = filter_data.copy()
당일_등락률['당일 등락률'] = (filter_data['일봉종가'] - filter_data['일봉시가'])/filter_data['일봉시가']
당일_등락률 = 당일_등락률.filter(like  = '등락률',axis= 1)
당일_등락률.head()
```

# 일주일 후의 등락률 최댓값(공시날의 저가와 일주일 후의 고가를 이용하여)

```python
일주후_등락률 = filter_data.copy()
일주후_등락률['일주후 등락률(저)'] = (filter_data['일주후종가'] - filter_data['일봉저가'])/filter_data['일봉저가']
일주후_등락률 = 일주후_등락률.filter(like  = '등락률',axis= 1)
일주후_등락률.head()
```

## 공시 당일날 종가 일주후 종가 비교

```python
일주후_등락률_종 = filter_data.copy()
일주후_등락률_종['일주후 등락률(종)'] = (filter_data['일주후종가'] - filter_data['일봉종가'])/filter_data['일봉종가']
일주후_등락률_종 = 일주후_등락률_종.filter(like  = '등락률',axis= 1)
일주후_등락률_종.head()
```

# 공시당일날 시가 일주일 후 종가 비교

```python
일주후_등락률_종 = filter_data.copy()
일주후_등락률_종['일주후 등락률(시)'] = (filter_data['일주후종가'] - filter_data['일봉시가'])/filter_data['일봉시가']
일주후_등락률_종 = 일주후_등락률_종.filter(like  = '등락률',axis= 1)
일주후_등락률_종.head()
```
