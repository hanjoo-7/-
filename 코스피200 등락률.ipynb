```python
import pandas as pd
import numpy as np
```

```python
#코스피200 항목들 불러오기
코스피200 = pd.read_csv('KOSPI200.csv', encoding='euc-kr')
코스피200.head()
```

```python
#불러온 종목 코드들을 isin()을 사용하기 위해 리스트형으로 변환 
kospi200 = list(np.array(코스피200['코드'].tolist()))
kospi200
```

```python
#투자자 엑셀 파일 불러오기
투자자 = pd.read_csv('투자자20201103.csv', encoding='euc-kr')
투자자.head()
```

```python
#kospi200 항목들만을 분류해주기
투자자 = 투자자[투자자['종목'].isin(kospi200)]
투자자
```

```python
#기존코드
#조건에 따라 분류해주기 
기관매수 = 투자자[
            (
                ((투자자.기관순매수1>0) & (투자자.기관순매수2>0)) |
                ((투자자.기관순매수2>0) & (투자자.기관순매수3>0)) |
                ((투자자.기관순매수1>0) & (투자자.기관순매수3>0))
            ) |
            ((투자자.기관순매수1>0) & (투자자.기관순매수2>0) & (투자자.기관순매수3>0))]
기관매수
```

##### 설정해주는 조건에 따라 2~3일 연속으로 매수한 경우 

```python
#s는 매수하는 대상 

s = '기관'

#SettingWithCopyWarning: 에러가 발생하여 .copy를 추가해줌
순매수 = 투자자[
    (
                ((투자자.filter(regex = '^{}순매수1$'.format(s))['{}순매수1'.format(s)]>0) & 
                 (투자자.filter(regex = '^{}순매수2$'.format(s))['{}순매수2'.format(s)]>0))
                 |
                ((투자자.filter(regex = '^{}순매수2$'.format(s))['{}순매수2'.format(s)]>0) & 
                 (투자자.filter(regex = '^{}순매수3$'.format(s))['{}순매수3'.format(s)]>0))
                 |
                ((투자자.filter(regex = '^{}순매수1$'.format(s))['{}순매수1'.format(s)]>0) & 
                 (투자자.filter(regex = '^{}순매수3$'.format(s))['{}순매수3'.format(s)]>0))
                 |
            ((투자자.filter(regex = '^{}순매수1$'.format(s))['{}순매수1'.format(s)]>0) &
             (투자자.filter(regex = '^{}순매수2$'.format(s))['{}순매수2'.format(s)]>0) & 
             (투자자.filter(regex = '^{}순매수3$'.format(s))['{}순매수3'.format(s)]>0))
    )].copy()

#투자자유형에 따라
#등락률을 넣어준 이유는 이후에 filter를 사용할때 한번에 불러오기 용이하기 위해서 
순매수.loc[:,'등락률_유형'] = s
순매수.rename(columns = {'종목':'종목_등락률'},inplace = True)
순매수
```

```python
#매수하는 날짜로부터 10일 이내에 매도할 경우의 등락률 구해주기.
#s_date는 매수 날짜 
s_date = 3

for i in range(10-s_date):
    순매수['등락률'+ str(s_date)+'매수' + str(10-i) +'매도'] = (순매수['종가'+ str(10-i)] - 순매수['종가' + str(s_date)]) / 순매수['종가'+ str(s_date)]
```

```python
순매수
```

```python
#등락률 포함하는 컬럼만 뽑아주기 (위의 경우 종목 코드와 투자자 유형 그리고 등락률)
등락률 = 순매수.filter(like  = '등락률',axis= 1)
```

### 종목별로 구해줄 경우

```python
#종목별 등락률의 평균을 구해줄 경우 
종목별_평균 = 등락률.groupby('종목_등락률',as_index = False).mean()
종목별_평균['투자자 유형'] = 등락률.iloc[0]['등락률_유형']

#컬럼 이름을 다시 원래대로 설정 
종목별_평균.rename(columns = {'종목_등락률':'종목 코드'},inplace = True)

#투자자 유형과 종목 코드를 인덱스로 설정 
종목별_평균 = 종목별_평균.set_index(['투자자 유형','종목 코드'])
종목별_평균
```

```python
#최대평균등락률 구해주기
#추후에 행전체를 추출하는 코드 추가 예정 
종목별_평균.idxmax()
```

```python
#최소평균등락률 구해주기
종목별_평균.idxmin()
```

### 종목이 아닌 전체의 등락률을 구하는 경우 

```python
average = pd.DataFrame(columns = ['매수/매도','평균 등락률'])

#종목코드는 필요가 없으므로 
등락률 = 등락률.filter(regex = '매도$')

for each in 등락률.columns:
    average = average.append(pd.DataFrame([[each,등락률[each].mean()]],columns = ['매수/매도','평균 등락률']),ignore_index = True)
    
print(average['평균 등락률'].max())
average
```
