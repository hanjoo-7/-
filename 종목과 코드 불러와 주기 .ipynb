```python
from bs4 import BeautifulSoup
import csv 
import os
import re
import requests 
import pandas as pd
import numpy as np
```

# 코스피200 지수 항목들 크롤링 이후 저장 

```python
#기존 코드의 경우 엑셀 파일을 다시 불러올시 삼성전자가 컬럼으로 들어가 버림. 
#기존 코드와 다르게 반복문 안에서 일일이 적는것이 아니라 데이터프레임을 csv로 전환

BaseUrl = "http://finance.naver.com/sise/entryJongmok.nhn?&page="

#기존에 파일이 존재할 경우 제거 
if os.path.exists("KOSPI200.csv"):
    os.remove("KOSPI200.csv")
else:
    pass

#종목들의 코드와 이름을 저장할 리스트 선언
code_list = []
name_list = []

for i in range(1,22,1):
    try:
        url = BaseUrl + str(i)
        r = requests.get(url)
        soup = BeautifulSoup(r.text,'lxml')
        items = soup.find_all('td',{'class':'ctg'})
        
        for item in items:
            txt = item.a.get('href')
            k = re.search('[\d]+',txt)
            if k:
                code_list.append(k.group())
                name_list.append(item.text) 
    except:
        pass
    
data = {"이름": name_list,"코드" : code_list}
data = pd.DataFrame(data)
data.to_csv("KOSPI200.csv", encoding='euc-kr')
```
